```python
import pandas as pd
import numpy as np
```

# Load and Process Data

## Channel Flow

```python
mean_channel_df = pd.read_csv('/Users/alexis/Cranfield/Group Project/Repos1/data/oden_db/channel/mean_combined_channel.csv')
vel_fluc_channel_df = pd.read_csv('/Users/alexis/Cranfield/Group Project/Repos1/data/oden_db/channel/vel_fluc_combined_channel.csv')
vp_channel_df = pd.read_csv('/Users/alexis/Cranfield/Group Project/Repos1/data/oden_db/channel/vp_combined_channel.csv')
```

mean_channel_df

```python
print(mean_channel_df.shape)
mean_channel_df.head()
```

```python
import math

# Convert 'Lx' and 'Lz' columns
mean_channel_df['Lx'] = mean_channel_df['Lx'].apply(lambda x: eval(x.replace('pi', f' * math.pi')))
mean_channel_df['Lz'] = mean_channel_df['Lz'].apply(lambda x: eval(x.replace('pi', f' * math.pi')))
```

vp_channel_df

```python
print(vp_channel_df.shape)
vp_channel_df.head()
```

```python
# Convert 'Lx' and 'Lz' columns
vp_channel_df['Lx'] = vp_channel_df['Lx'].apply(lambda x: eval(x.replace('pi', f' * math.pi')))
vp_channel_df['Lz'] = vp_channel_df['Lz'].apply(lambda x: eval(x.replace('pi', f' * math.pi')))
```

## Couette Flow

```python
mean_couette_df = pd.read_csv('/Users/alexis/Cranfield/Group Project/Repos1/data/oden_db/couette/mean_combined_couette.csv')
vel_fluc_couette_df = pd.read_csv('/Users/alexis/Cranfield/Group Project/Repos1/data/oden_db/couette/vel_fluc_combined_couette.csv')
vp_couette_df = pd.read_csv('/Users/alexis/Cranfield/Group Project/Repos1/data/oden_db/couette/vp_combined_couette.csv')
```

mean_couette_df

```python
print(mean_couette_df.shape)
mean_couette_df.head()
```

```python
# Convert 'Lx' and 'Lz' columns
mean_couette_df['Lx'] = mean_couette_df['Lx'].apply(lambda x: eval(x.replace('pi', f' * math.pi')))
mean_couette_df['Lz'] = mean_couette_df['Lz'].apply(lambda x: eval(x.replace('pi', f' * math.pi')))
```

```python
# Remove '(+/-)' from 'U_mean' column
mean_couette_df['U_mean'] = mean_couette_df['U_mean'].apply(lambda x: x.replace('(+/-)', ''))
```

vel_fluc_couette_df

```python
print(vel_fluc_couette_df.shape)
vel_fluc_couette_df.head()
```

```python
# Convert 'Lx' and 'Lz' columns
vel_fluc_couette_df['Lx'] = vel_fluc_couette_df['Lx'].apply(lambda x: eval(x.replace('pi', f' * math.pi')))
vel_fluc_couette_df['Lz'] = vel_fluc_couette_df['Lz'].apply(lambda x: eval(x.replace('pi', f' * math.pi')))
```

```python
# Remove '(+/-)' from 'U_mean' column
vel_fluc_couette_df['U_mean'] = vel_fluc_couette_df['U_mean'].apply(lambda x: x.replace('(+/-)', ''))
```

vp_couette_df

```python
print(vp_couette_df.shape)
vp_couette_df.head()
```

```python
# Convert 'Lx' and 'Lz' columns
vp_couette_df['Lx'] = vp_couette_df['Lx'].apply(lambda x: eval(x.replace('pi', f' * math.pi')))
vp_couette_df['Lz'] = vp_couette_df['Lz'].apply(lambda x: eval(x.replace('pi', f' * math.pi')))
```

```python
# Remove '(+/-)' from 'U_mean' column
vp_couette_df['U_mean'] = vp_couette_df['U_mean'].apply(lambda x: x.replace('(+/-)', ''))
```

# Merge Data

## Channel Flow

```python
df_combined_channel = pd.merge(
    mean_channel_df,
    vp_channel_df,
    on=["nx", "ny", "nz", "Lx", "Lz", "n", "nu", "delta", "u_tau", "y/delta", "y^+", "Re_tau", "U_mean"]
)

df_combined_channel = pd.merge(
    df_combined_channel,
    vel_fluc_channel_df,
    on=["nx", "ny", "nz", "Lx", "Lz", "n", "nu", "delta", "u_tau", "y/delta", "y^+", "Re_tau", "U_mean"]
)
```

```python
df_combined_channel.columns
```

```python
df_combined_channel.to_csv('/Users/alexis/Cranfield/Group Project/Repos1/data/oden_db/channel/combined_channel.csv', index=False)   
```

## Couette Flow

```python
df_combined_couette = pd.merge(
    mean_couette_df,
    vel_fluc_couette_df,
    on=["nx", "ny", "nz", "Lx", "Lz", "n", "nu", "delta", "u_tau", "y/delta", "y^+", "Re_tau", "U_mean"]
)

df_combined_couette = pd.merge(
    df_combined_couette,
    vp_couette_df,
    on=["nx", "ny", "nz", "Lx", "Lz", "n", "nu", "delta", "u_tau", "y/delta", "y^+", "Re_tau", "U_mean"]
)
```

```python
df_combined_couette.columns
```

```python
df_combined_couette.to_csv('/Users/alexis/Cranfield/Group Project/Repos1/data/oden_db/couette/combined_couette.csv', index=False)
```

# Concatenate Data

```python
df_combined = pd.concat([df_combined_channel, df_combined_couette])
```
