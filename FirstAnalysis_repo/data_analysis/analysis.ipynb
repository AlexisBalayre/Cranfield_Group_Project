# Exemple of little data analysis for parameter ww 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('../data/oden_db/channel/ww_combined.csv')
df.head()
```

Analysis 

```python

# 1. Line Plot of Turbulent Transport vs. y/delta for Different Re_tau
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="y/delta", y="Turbulent_Transport",
             hue="Re_tau")
plt.title("Turbulent Transport vs. y/delta for Different Re_tau")
plt.xlabel("y/delta")
plt.ylabel("Turbulent Transport")
plt.legend(title="Re_tau")
plt.show()

# 2. Scatter Plot of Viscous Dissipation vs. y^+ for Different Re_tau
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="y^+", y="Viscous_Dissipation",
                hue="Re_tau", alpha=0.7)
plt.title("Viscous Dissipation vs. y^+ for Different Re_tau")
plt.xlabel("y^+")
plt.ylabel("Viscous Dissipation")
plt.legend(title="Re_tau")
plt.show()

# 3. Box Plot of Pressure Strain for Different Re_tau
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="Re_tau", y="Pressure_Strain")
plt.title("Pressure Strain for Different Re_tau")
plt.xlabel("Re_tau")
plt.ylabel("Pressure Strain")
plt.show()
```

```python

# 4. Heatmap of Correlation Between Variables for a Specific Re_tau here : 5200
df_specific_re_tau = df[df['Re_tau'] == 5200]

# Compute the correlation matrix, excluding the 'Re_tau' column
corr = df_specific_re_tau.drop('Re_tau', axis=1).corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix Heatmap for Re_tau = 5200")
plt.show()
```
