```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Analysis of Channel Flow Data from CFD_simulation simulation

```python
channel_data = pd.read_csv('../../data/CFD_simulation/channel_test.csv') # Channel Flow Test Data
k_mu_k_epsilon = pd.read_csv('../../data/CFD_simulation/k_mu_k_epsilon.csv') # k and mu_t (K-epsilon model)
k_mu_k_omega = pd.read_csv('../../data/CFD_simulation/k_mu_k_omega.csv') # k and mu_t (K-omega model)
```

### Head Values

```python
print("Shape of channel_data: ", channel_data.shape)
channel_data.head()
```

```python
print("Shape of k_mu_k_epsilon: ", k_mu_k_epsilon.shape)
k_mu_k_epsilon.head()
```

```python
print("shape of k_mu_k_epsilon: ", k_mu_k_epsilon.shape)
k_mu_k_omega.head()
```

### Correlation Analysis

```python
channel_data_corr = channel_data.corr()**2

plt.figure(figsize=(12, 8))
plt.title('Correlation Matrix of Channel Data')
sns.heatmap(channel_data_corr, annot=True, cmap='magma', linewidths=0.5)
plt.show()
```

```python
k_mu_k_epsilon_corr = k_mu_k_epsilon.corr()**2

plt.figure(figsize=(12, 8))
plt.title('Correlation Matrix of k_mu_k_epsilon Data')
sns.heatmap(k_mu_k_epsilon_corr, annot=True, cmap='magma', linewidths=0.5)
plt.show()
```

```python
k_mu_k_omega_corr = k_mu_k_omega.corr()**2

plt.figure(figsize=(12, 8))
plt.title('Correlation Matrix of k_mu_k_omega Data')
sns.heatmap(k_mu_k_omega_corr, annot=True, cmap='magma', linewidths=0.5)
plt.show()
```
