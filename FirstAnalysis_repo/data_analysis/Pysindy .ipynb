# Importing libraries

```python
import numpy as np
import pysindy as ps
from scipy.integrate import solve_ivp
import pandas as pd
```

```python
df_pca_k = pd.read_csv("data_PCA_k.csv")
df_pca_g = pd.read_csv("data_PCA_gen.csv")
df_tsne = pd.read_csv("data_tsne.csv")
print(df_pca_k)
print(df_pca_g)
print(df_tsne)
```

## PySindy

```python
X = df_pca_k.values
t = np.arange(0, len(df_pca_k)) * 0.01
model = ps.SINDy()
model.fit(X, t=t)
model.print()
```
